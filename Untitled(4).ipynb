```python
# quantile 
q1 = combinedata.quantile(.25)
q3 = combinedata.quantile(.75)
```

```python
def find_outliers(combinedata,coln):
    q1 = combinedata[coln].quantile(0.25)
    q3 = combinedata[coln].quantile(0.75)
    iqr = q3 - q1
    min_r = q1 - 1.5 * iqr
    max_r = q3 + 1.5 * iqr
    outliers_indices = combinedata.index[(combinedata[coln] < min_r) | (combinedata[coln] > max_r)]
    return outliers_indices
```

```python
# finding outliers
find_outliers(combinedata,'cnt')
#outliers = combinedata[~((combinedata - combinedata.mean()).abs() > 3 * combinedata.std).any(axis=1)]
#combinedata
```
